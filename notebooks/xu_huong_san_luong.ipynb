```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Sản lượng tiềm năng

Sản lượng tiềm năng không cố định mà thường xuyên thay đổi theo thời gian. Sự thay đổi của sản lượng tiềm năng phụ thuộc vào sự thay đổi của số lượng và chất lượng nguồn tài nguyên. Thông thường thì lượng vốn, lượng lao động, năng suất lao động có khuynh hướnng tăng theo thời gian. Vì vậy, trừ những trường hợp đặc biệt như chiến tranh, thiên tai, ... nói chung sản lượng tiềm năng có xu hướng gia tăng theo thời gian.

```python
# make data
x = np.linspace(0, 10, 100)
y = math.tan(math.radians(30)) * x + 2

style = 'classic'
with plt.style.context(style):
    fig, ax = plt.subplots()
    ax.plot(x, y, linewidth=2.0)
    ax.set_xlim((0,10))
    ax.set_ylim((0,10))
    ax.set_xlabel("t", loc='right')
    ax.set_ylabel("Y", rotation=0, loc='top')
    ax.set_xticks([])
    ax.set_yticks([])
    plt.grid()
    plt.show()
```

# Sản lượng tiềm năng theo mức giá

Đồ thị của sản lượng tiềm năng theo mức giá:

Sản lượng tiềm năng không phụ thuộc vào mức giá mà phụ thuộc vào các nguồn lực của nền kinh tế. Do đó, đồ thị sản lượng tiềm năng theo mức giá có dạng sau:

```python
# make data
x = np.linspace(0, 10, 100)
y = math.tan(math.radians(90)) * x

style = 'classic'
with plt.style.context(style):
    fig, ax = plt.subplots(1,1,num=style)
    ax.plot(x, y, linewidth=2.0)
    ax.set_xlim((-5,5))
    ax.set_ylim((0,10))
    ax.set_xlabel(r'$Y_p$')
    ax.set_ylabel("P", rotation=0)
    ax.set_xticks([])
    ax.set_yticks([])
    plt.grid()
    plt.show()
```
